```python
import hvplot.pandas
import param
import panel as pn
import sys
sys.path.append(r"C:\Users\xavier.mouy\Documents\Workspace\GitHub\ecosound")
from ecosound.core.measurement import Measurement

data_file=r'C:\Users\xavier.mouy\Documents\PhD\Projects\Dectector\results\dataset_FS-NN.nc'

# load dataset
dataset = Measurement()
dataset.from_netcdf(data_file)
data = dataset.data

pn.extension()
inputs = list(data.columns)
    
class DataDashboard(param.Parameterized):
    X_variable = param.Selector(inputs, default=inputs[26])
    Y_variable = param.Selector(inputs, default=inputs[27])
    Color = param.Selector(inputs, default=inputs[33])
    alpha_slider = param.Number(0.5, bounds=(0, 1))
    #size_slider = param.Number(6, bounds=(1, 10))
    size_slider = param.Integer(6, bounds=(1, 20))
    
    @param.depends('X_variable', 'Y_variable','Color','alpha_slider','size_slider')
    def scatter_plot(self):
        scatter_plot = data.hvplot.scatter(x=self.X_variable,
                                           y=self.Y_variable,
                                           s=self.size_slider,
                                           by=self.Color,
                                           alpha=self.alpha_slider,
                                           #datashade=True,
                                           #dynspread=True,
                                           #rasterize=False,
                                           #aggregator='count',
                                           hover=True,
                                           legend=True,
                                          )
        scatter_plot.opts(click_policy="hide")
        #scatter_plot.legend.click_policy="hide"
        return scatter_plot   
    
    @param.depends('X_variable', 'Color','alpha_slider')
    def hist_plot(self):
        #columns = ['Violent Crime rate', 'Property crime rate', 'Burglary rate']
        hist_plot = data.hvplot.hist(y=self.X_variable,
                                     by=self.Color,
                                     bins=50,
                                     alpha=0.5,
                                     legend='right',
                                     )
        return hist_plot
    
    def panel(self):
        #C1 = pn.Column(self.scatter_plot,self.hist_plot)
        #C2 = pn.Row(self.param,C1)
        C2 = pn.Row(self.param,self.scatter_plot)
        return C2
dashboard = DataDashboard(name='Dataset_Dashboard')

dashboard.panel().servable()
```
